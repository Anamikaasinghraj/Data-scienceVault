```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x, y)
```

# adding labels

```python
plt.title("xy graphs")
plt.xlabel(" x axis")
plt.ylabel("y axis")
plt.plot(x,y)
```

## subplot

```python
plt.subplot(2,2,1)
plt.plot(y,y)
plt.subplot(2,2,3)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(y,x)
```

# object oriented way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.4])
axis2.plot(x,y)
```
