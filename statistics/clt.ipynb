```python
import numpy as np
import matplotlib.pyplot as plt
```

### Chapter 8: Central Limit Theorem

#### CLT:- The Central Limit Theorem (CLT) says that the average of a large number of independent, random samples from any population will be approximately normally distributedâ€”even if the original data isn't. This makes it easier to use normal distribution tools for analysis, especially in hypothesis testing and confidence intervals

```python
# Generate a non-normal population
population = np.random.exponential(scale=2, size=10000)

# Draw many samples and compute means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot population vs sampling distribution
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Population (Exponential)')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title('Sampling Distribution of the Mean (n=30)')

plt.show()
```
