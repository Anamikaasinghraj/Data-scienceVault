### Chapter 7: Probability Distributions

#### Normal distribution: A normal distribution is a way to describe how data is spread out. It looks like a smooth, bell-shaped curve where:
##### Most values are close to the average (called the mean)
##### Fewer values are found as you move farther from the average
##### The curve is symmetrical, meaning both sides look the same

```python
mean = 0
std_dev = 1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(8,4))
plt.plot(x, y, label='Normal Distribution')
plt.title("Normal Distribution Curve")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

#### binomial distribution:- A binomial distribution describes the number of times a specific outcome happens in a fixed number of tries, when each try has only two possible results: success or failure.

```python
# Binomial Distribution
n = 10  # trials
p = 0.5  # success probability
x = np.arange(0, n+1)
binom_probs = binom.pmf(x, n, p)

plt.figure(figsize=(8,4))
plt.bar(x, binom_probs, color='purple')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

#### poisson distribution:-  A Poisson distribution is used to model the number of times an event happens in a fixed interval of time or space, when the events happen randomly and independently.

```python
mu = 3
x = np.arange(0, 10)
poisson_probs = poisson.pmf(x, mu)

plt.figure(figsize=(8,4))
plt.bar(x, poisson_probs, color='orange')
plt.title("Poisson Distribution (mu=3)")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
